```python
import pandas as pd
import numpy as np
from tqdm import tqdm, tqdm_notebook
```

```python
offline_test = pd.read_csv(r"C:\Users\IE\Desktop\ict3\ccf_offline_stage1_test_revised.csv", dtype=str, parse_dates=['Date_received'])
offline_train = pd.read_csv(r"C:\Users\IE\Desktop\ict3\ccf_offline_stage1_train.csv", dtype=str, parse_dates=['Date_received', 'Date'])
online_train = pd.read_csv(r"C:\Users\IE\Desktop\ict3\ccf_online_stage1_train.csv", dtype=str, parse_dates=['Date_received', 'Date'])
```

```python
def sperate_offline(df):
    buy = df[pd.isnull(df['Coupon_id'])]
    received = df[(~pd.isnull(df['Coupon_id'])) & (pd.isnull(df['Date']))]
    ps = df[(~pd.isnull(df['Coupon_id'])) & (~pd.isnull(df['Date'])) \
                      & ((df['Date']-df['Date_received']).apply(lambda x:x.days) <= 15)]
    ns = df[(~pd.isnull(df['Coupon_id'])) & (~pd.isnull(df['Date'])) \
                      & ((df['Date']-df['Date_received']).apply(lambda x:x.days) > 15)]
    
    return buy, received, ps, ns

buy, received, ps, ns = sperate_offline(offline_train)
```

```python
def sperate_online(df):
    click = df[df['Action'] == '0']
    get_coupon = df[df['Action'] == '2']
    tmp = df[(df['Action'] == '1')]
    buy = df[(df['Action'] == '1') & (pd.isnull(df['Coupon_id']))]
    ps = df[(df['Action'] == '1') & (~pd.isnull(df['Coupon_id'])) \
                      & ((df['Date']-df['Date_received']).apply(lambda x:x.days) <= 15)]
    ns = df[(df['Action'] == '1') & (~pd.isnull(df['Coupon_id'])) \
                      & ((df['Date']-df['Date_received']).apply(lambda x:x.days) > 15)]
    
    return get_coupon, click, buy, ps, ns

get_coupon, click, buy, ps, ns = sperate_online(online_train)
```

```python
def sperate_online_userID(user_id, df):
    
    tmp = df[df['User_id'] == user_id]
    if len(tmp) != 0:
        get_coupon, click, buy, ps, ns = sperate_online(tmp)
        tmp_nan = [user_id, np.nan, len(get_coupon), len(click), len(buy), len(ps), len(ns)]
    else:
        tmp_nan ==[user_id, np.nan, 0, 0, 0, 0, 0]

    return tmp_nan
```

```python
tmp = (online_train['Date']-online_train['Date_received']).apply(lambda x:x.days)
tmp[:10]
```

```python
userIDs = online_train['User_id'].unique()
res = pd.DataFrame(columns = ['User_id'], data=userIDs)
res['Discount_rate'] = 0
res['get_coupon'] = 0
res['click'] = 0
res['buy'] = 0
res['ps'] = 0
res['ns'] = 0

res.set_index('User_id', inplace=True)
for _, i in tqdm_notebook(online_train.iterrows(), total=len(online_train)):
    index = i['User_id']
    if i['Action'] == '0':
        res.loc[index, 'click'] += 1
    elif i['Action'] == '2':
        res.loc[index, 'get_coupon'] += 1
    elif i['Action'] == '1':
        if pd.isnull(i['Coupon_id']):
            res.loc[index, 'buy'] += 1
        else:
            if tmp[_] <= 15:
                res.loc[index, 'ps'] += 1
            elif tmp[_] > 15:
                res.loc[index, 'ns'] += 1
res = res.reset_index()
```
